```python
df_silver = spark.table("ecommerce_electroics_sales.sales_schema.silver_sales")
```

```python
from pyspark.sql.functions import sum as spark_sum, col,count,round

df_gold_1 = df_silver.groupBy("Month")\
                     .agg(round(spark_sum("Revenue"),2).alias("monthly_Revenue"))\
                     .orderBy(col("Month"))

df_gold_1.show()
```

```python
df_gold_1.write.format("delta").mode("overwrite").saveAsTable("ecommerce_electroics_sales.sales_schema.gold_monthly_revenue")
```

```python
df_gold_2 = df_silver.groupBy("Product")\
                     .agg(spark_sum(col("Quantity_Ordered")).alias("Total_Quantity_Sold"),\
                         round(spark_sum(col("Revenue")),2).alias("Total_Revenue")).orderBy(col("Total_Quantity_Sold").desc())
df_gold_2.display()
```

```python
df_gold_2.write.format("delta").mode("overwrite").saveAsTable("ecommerce_electroics_sales.sales_schema.gold_product_sales")
```

```python
df_gold_3 = df_silver.groupBy("City").agg(count("Order_ID").alias("Total_Orders"),\
                                           round(spark_sum(col("Revenue")),2).alias("Total_Revenue")).orderBy(col("Total_Revenue").desc())
df_gold_3.display()
df_gold_3.write.format("delta").mode("overwrite").saveAsTable("ecommerce_electroics_sales.sales_schema.gold_city_sales")
```

```python
df_gold_4 = df_silver.groupBy("Hour").agg(count("Order_ID").alias("Total_Orders")).orderBy(col("Hour"))
df_gold_4.show()

df_gold_4.write.format("delta").mode("overwrite").saveAsTable("ecommerce_electroics_sales.sales_schema.gold_hourly_sales")
```
